```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
import numpy as np
import cv2
import matplotlib.pyplot as plt
from easygui import *
import os
from PIL import Image, ImageTk
from itertools import count
import tkinter as tk
import string
import speech_recognition as sr


class ImageLabel(tk.Label):
        def load(self, im):
            
            if isinstance(im, str):
            
                im = Image.open(im)
            
            self.loc = 0
            self.frames = []
            
            try:
                for i in count(1):
                    self.frames.append(ImageTk.PhotoImage(im.copy()))
                    im.seek(i)
            except EOFError:
                pass

            try:
                self.delay = im.info['duration']
            except:
                self.delay = 100

            if len(self.frames) == 1:
                self.config(image=self.frames[0])
            else:
                self.next_frame()

        def unload(self):
            self.config(image=None)
            self.frames = None

        def next_frame(self):
            if self.frames:
                self.loc += 1
                self.loc %= len(self.frames)
                self.config(image=self.frames[self.loc])
                self.after(self.delay, self.next_frame)


def isl():

    # create a Recognizer object
    r = sr.Recognizer()

    # ask user to choose input method
    input_method = input("Choose input method:\n1. Speech\n2. Text")

    while True:

        # use the selected input method
        flag = True
        if input_method == "1":
            # use the microphone as the source of input
            with sr.Microphone() as source:
                print("Speak something...")
                # listen for audio input
                audio = r.listen(source)

            # use Google speech recognition to translate audio to text
            try:
                text = r.recognize_google(audio)
                print("You said: {}".format(text))
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand the audio.")
                flag = False
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
                flag = False
        elif input_method == "2":
            # ask user to input text
            text = input("Enter text: ")
            print("text: {}".format(text))
        else:
            print("Invalid input method selected.")
            flag = False
        
        if flag == False or len(text) == 0:
            continue


        if ' ' in text:  # If the input text contains space, it is a sentence
            # Tokenize the text
            tokens = nltk.word_tokenize(text)
            
            
            stopwords = ['a', 'an', 'is','are','were','if','it','be','ing','also','ed','the', 'and','was', 'or', 'but', 'not', 'for', 'of', 'on', 'in', 'to', 'at', 'by', 'with', 'from']
            words = text.split()
            filtered_words = []
            for word in words:
                if word.lower() not in stopwords:
                    filtered_words.append(word)
                    filtered_sentence = ' '.join(filtered_words)

            def reorder_sentence(sentence):
                words = sentence.split()  # Split the sentence into words
                subject = words.pop(0)  # Get the first word as the subject and remove it from the list
                verb = words.pop(0)  # Get the next word as the verb and remove it from the list
                object = " ".join(words)  # Combine the remaining words as the object
                
                # Concatenate the subject, object, and verb in the order of subject-object-verb
                reordered_sentence = subject + " " + object + " " + verb
                
                return reordered_sentence

            reordered_sentence = reorder_sentence(filtered_sentence)
            
            

        
            # Create a PorterStemmer object
            stemmer = PorterStemmer()

            # Stem each word in the filtered sentence
            stemmed_words = [stemmer.stem(word) for word in reordered_sentence.split()]

            # Join the stemmed words back into a sentence
            stemmed_sentence = " ".join(stemmed_words)

            # Create a WordNetLemmatizer object
            lemmatizer = WordNetLemmatizer()

            # Lemmatize each word in the stemmed sentence
            lemmatized_words = [lemmatizer.lemmatize(word) for word in stemmed_sentence.split()]

            # Join the lemmatized words back into a sentence
            lemmatized_sentence = " ".join(lemmatized_words)

            # Print the results
            print("Input text: {}".format(text))
            print("Tokenized text:", tokens)
            print("Filtered sentence: {}".format(filtered_sentence))
            print("Reordered sentence: {}".format(reordered_sentence))
            print("Stemmed sentence: {}".format(stemmed_sentence))
            print("Lemmatized sentence: {}".format(lemmatized_sentence))
            
            inputAs = lemmatized_sentence

        else:  # If the input text does not contain space, it is a one-word input
            # Create a PorterStemmer object
            stemmer = PorterStemmer()

            # Stem the input word
            stemmed_word = stemmer.stem(text)

            # Create a WordNetLemmatizer object
            lemmatizer = WordNetLemmatizer()

            # Lemmatize the stemmed word
            lemmatized_word = lemmatizer.lemmatize(stemmed_word)

            # Print the results
            print("Input text: {}".format(text))
            print("Stemmed word: {}".format(stemmed_word))
            print("Lemmatized word: {}".format(lemmatized_word))


            inputAs = lemmatized_word

            
        #----------------------------------------------------------------------------------------------------------------------

        path = "C://Users//USER//Desktop//program and Gif//isl_gif1"
        isl_gif = os.listdir(path)

        #--------------------------------------------------------------------------------------------------------------------
        
        
        arr=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
        
        i=0
        try:
                a = inputAs
                for c in string.punctuation:
                    a= a.replace(c,"")
                    
                if(a.lower()=='goodbye' or a.lower()=='good bye' or a.lower()=='bye' or a.lower()=='quit'):
                        print("oops!Time To say good bye")
                        break
                
                elif(a.lower()+".gif" in isl_gif):
                
                    
                    # class cutted from here

                    root = tk.Tk()
                
                    lbl = ImageLabel(root)
                
                    lbl.pack()
                
                    lbl.load(r'isl_gif1/{0}.gif'.format(a.lower()))
                    
                    root.after(3000,lambda:root.destroy())
                
                    root.mainloop()
                
            
                
                else:

                    for word in a.lower().split():
                        if word+".gif" in isl_gif:
                            root = tk.Tk()
                
                            lbl = ImageLabel(root)
                        
                            lbl.pack()
                        
                            lbl.load(r'isl_gif1/{0}.gif'.format(word))
                            
                            root.after(3000,lambda:root.destroy())
                        
                            root.mainloop()

                        else:
                    #else part
                            for i in range(len(word)):
                                            if(word[i] in arr):
                                    
                                                    ImageAddress = 'letters/'+word[i]+'.jpg'
                                                    ImageItself = Image.open(ImageAddress)
                                                    ImageNumpyFormat = np.asarray(ImageItself)
                                                    plt.imshow(ImageNumpyFormat)
                                                    plt.draw()
                                                    plt.pause(0.8)
                                            else:
                                                    continue
                    #-----------

        except:
                print(" ")
        plt.close()

def asl():

    # create a Recognizer object
    r = sr.Recognizer()

    # ask user to choose input method
    print("Choose input method:\n1. Speech\n2. Text")
    input_method = input()

    while True:

        # use the selected input method
        flag = True
        if input_method == "1":
            # use the microphone as the source of input
            with sr.Microphone() as source:
                print("Speak something...")
                # listen for audio input
                audio = r.listen(source)

            # use Google speech recognition to translate audio to text
            try:
                text = r.recognize_google(audio)
                print("You said: {}".format(text))
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand the audio.")
                flag = False
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
                flag = False
        elif input_method == "2":
            # ask user to input text
            text = input("Enter text: ")
            print("text: {}".format(text))
        else:
            print("Invalid input method selected.")
            flag = False
        
        if flag == False or len(text) == 0:
            continue

        #----------------------------------------------------------------------------------------------------------------------

        path = "C://Users//USER//Desktop//program and Gif//ASL_GIF"
        isl_gif = os.listdir(path)

        #--------------------------------------------------------------------------------------------------------------------
        
        
        arr=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

        i=0
        try:
                
                a = text
                for c in string.punctuation:
                    a= a.replace(c,"")
                    
                if(a.lower()=='goodbye' or a.lower()=='good bye' or a.lower()=='bye' or a.lower()=='quit'):
                        print("oops!Time To say good bye")
                        break
                
                elif(a.lower()+".gif" in isl_gif):

                    root = tk.Tk()
                
                    lbl = ImageLabel(root)
                
                    lbl.pack()
                
                    lbl.load(r'ASL_GIF/{0}.gif'.format(a.lower()))

                    root.after(6000,lambda:root.destroy())
                
                    root.mainloop()
            
                
                else:
        
                    for i in range(len(a)):
                                    if(a[i] in arr):
                            
                                            ImageAddress = 'letters/'+a[i]+'.jpg'
                                            ImageItself = Image.open(ImageAddress)
                                            ImageNumpyFormat = np.asarray(ImageItself)
                                            plt.imshow(ImageNumpyFormat)
                                            plt.draw()
                                            plt.pause(0.8)
                                    else:
                                            continue

        except:
                print(" ")
        plt.close()



while True:
    choice = input("Choose option:\n1. ISL\n2. ASL\n3. Exit\n")
    if choice == "1":
        isl()
    elif choice == "2":
        asl()
    else:
        break
```
